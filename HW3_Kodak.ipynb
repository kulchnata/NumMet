```python
import numpy as np

A = np.array([[1/np.sqrt(2), -1/np.sqrt(2)],
              [1/2, 1/2],
              [-1/2, -1/2]])
b = np.array([1, 0, 1])

U, S, VT = np.linalg.svd(A, full_matrices=False)
np.linalg.pinv(np.diag(S))
x = VT.T @ np.linalg.pinv(np.diag(S)) @ U.T @ b

print("LSS:", x)
```

```python
C = pd.read_hdf(hdfName, 'C', mode = 'r+')
C
```

```python
def mat2vec(A):
    A = np.flipud(A)
    a = np.reshape(A, np.prod(A.shape))
    return a

def vec2mat(a, shape):
    A = np.reshape(a, shape)
    A = np.flipud(A)
    return A

plt.imshow(A, cmap='gray')
plt.title('Blurred Image:')
```

```python
C_inv = np.linalg.pinv(C)
a = mat2vec(A)

a0 = np.dot(C_inv, a)
A0_recovered = vec2mat(a0, (16, 51)) 

plt.imshow(A0_recovered, cmap='gray')
plt.title('Original Image:')
```

```python
import numpy as np
import matplotlib.pyplot as plt

U, Sigma, Vt = np.linalg.svd(C, full_matrices=False)
fraction_to_keep = 0.9
num_singular_values = int(fraction_to_keep * len(Sigma))


U_truncated = U[:, :num_singular_values]
Sigma_truncated = np.diag(Sigma[:num_singular_values])
Vt_truncated = Vt[:num_singular_values, :]

C_pseudo_inv_truncated = Vt_truncated.T @ np.linalg.inv(Sigma_truncated) @ U_truncated.T

a = mat2vec(A)

a0 = np.dot(C_pseudo_inv_truncated, a)

A0_recovered = vec2mat(a0, (16, 51))

plt.imshow(A0_recovered, cmap='gray')
plt.show()
```

```python
import numpy as np
import pandas as pd
hdfName = "localization_data_1.hdf"
pd.HDFStore(hdfName).keys()
```

```python
dr = pd.read_hdf(hdfName, 'dr', mode = 'r+')

measurements = pd.read_hdf(hdfName, 'measurements', mode = 'r+')

r = pd.read_hdf(hdfName, 'r', mode = 'r+')

dr.to_numpy()
```

```python
import sympy as s
import numpy as np

n = 3

def create_symbols(prefix, n):
    return [s.symbols(f"{prefix}{i}") for i in range(1, n + 1)]

def create_point_matrices(x_symbols, y_symbols):
    return [s.Matrix([x, y]) for x, y in zip(x_symbols, y_symbols)]

def create_corrections(dx_symbols, dy_symbols):
    return [s.Matrix([dx, dy]) for dx, dy in zip(dx_symbols, dy_symbols)]

def create_angle_expression(points):
    return acos((points[-1] - points[0]).dot(points[1] - points[0]) /(s.sqrt((points[-1] - points[0]).dot(points[-1] - points[0])) * s.sqrt((points[1] - points[0]).dot(points[1] - points[0]))))

x_symbols = create_symbols("x", n)
y_symbols = create_symbols("y", n)
dx_symbols = create_symbols("dx", n)
dy_symbols = create_symbols("dy", n)

points = create_point_matrices(x_symbols, y_symbols)
corrections = create_corrections(dx_symbols, dy_symbols)
updated_points = [correction + point for correction, point in zip(corrections, points)]

angle_exprt = create_angle_expression(updated_points)
angle_exprt
```

```python
var = tuple(x for pair in zip(x_symbols, y_symbols) for x in pair)
print(var)
subs = [pair for dx, dy in zip(dx_symbols, dy_symbols) for pair in [(dx, 0), (dy, 0)]]
print(subs)
angle_exprt.subs(subs)
```

```python
dang = [ diff(angle_exprt, v).subs(subs) for v in var ]
ang_f = s.lambdify(var, angle_exprt.subs(subs))
dang_f = s.lambdify(var, dang)
```

```python
b = [ [ measurements.to_numpy()[0][3] - ang_f(*r.to_numpy().reshape(6)) ] ]
A = [ dang_f(*r.to_numpy().reshape(6)) ]
np.linalg.lstsq(A, b)[0].reshape(3,2)
```
